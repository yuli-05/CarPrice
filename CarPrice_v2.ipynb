# **Preparación de datos**

```python
cars.head()
```

Antes de crear el modelo, debemos realizar algunos cambios en la columna 'CarName', debemos considerar solo el nombre del nombre de la empresa, no el nombre del modelo del automóvil.

```python
# Cambiar el nombre de los valores de la columna 'CarName' por el nombre de la empresa

cars['CarName']=cars['CarName'].apply(lambda x:x.split(" ")[0])
```

```python
cars.head()
```

## Creando variables dummy

```python
cars['symboling']=cars['symboling'].apply(lambda x:str(x))
```

```python
cars.info()
```

```python
# Creando dummies para variables categóricas

variables=['symboling','CarName','fueltype','aspiration','doornumber','carbody','drivewheel','enginelocation','enginetype',
          'cylindernumber','fuelsystem']

# Crear dummies para cada variable categórica y descartar las primeras variables ficticias para cada variable categórica

dummies=pd.get_dummies(cars[variables],drop_first=True)
```

```python
# Añadiendo estos dummies a nuestro conjunto de datos original

cars=pd.concat([cars,dummies],axis=1)
```

```python
# Revisemos nuestro conjunto de datos después de agregar estos dummies

cars.head()
```

```python
# Ahora descartando las variables categóricas, ya que hemos creado dummies y columnas derivadas para ellas y car_id.

categories=['car_ID','symboling','CarName','fueltype','aspiration','doornumber','carbody','drivewheel','enginelocation',
            'enginetype','cylindernumber','fuelsystem']

cars.drop(categories,axis=1,inplace=True)
```

```python
cars.head()
```

# **División de conjunto de datos**
En conjuntos de datos de entrenamiento y prueba.

```python
cars.info()
```

```python
# Importe las bibliotecas necesarias para realizar el primer paso de la regresión, es decir, dividir entrenar y probar

from sklearn.model_selection import train_test_split

cars_train,cars_test=train_test_split(cars,train_size=0.7,test_size=0.3,random_state=50)
```

## Escalando los conjuntos de datos
Aplicaremos el concepto de escala Min-Max para cambiar la escala del conjunto de datos del tren.

```python
# importar la biblioteca para escalar con MIN-MAX

from sklearn.preprocessing import MinMaxScaler
```

```python
# Creando objeto para MinMaxScaler

scaler=MinMaxScaler()
```

```python
# Ahora, aplicar el objeto escalar en el conjunto de datos del tren para reescalar

# Aplicaremos esto a todas las columnas excepto a las columnas dummy y a la columna de símbolos

scalling=['wheelbase','carheight','enginesize','boreratio','stroke','compressionratio','peakrpm','highwaympg','price']

cars_train[scalling]=scaler.fit_transform(cars_train[scalling])
```

```python
cars_train.describe()
```

Ahora, hemos escalonado todas las columnas y todas están en el rango entre 0 y 1

# **Construcción de un modelo lineal**

```python
# importar librerias necesarias

import statsmodels.api as sm
```

Dividir el conjunto de trenes en conjuntos X e Y para construir un modelo

```python
y_train=cars_train.pop('price')
X_train=cars_train
```

```python
# Y tenemos X_train 

X_train.head()
```

## Construyendo nuestro modelo usando RFE 
Eliminación de características recursivas.

```python
# Importación de RFE y regresión lineal

from sklearn.feature_selection import RFE
from sklearn.linear_model import LinearRegression

from statsmodels.stats.outliers_influence import variance_inflation_factor
```

```python
# Ejecutando RFE con una variable de salida igual a 10

lm=LinearRegression()   # Creando un objeto

lm.fit(X_train,y_train)

rfe=RFE(lm,20)          # Creando RFE con 20 variables

rfe=rfe.fit(X_train,y_train)
```

```python
list(zip(X_train.columns,rfe.support_,rfe.ranking_))  # Checking the rfe rank and rfe support
```

```python
# Creación de un conjunto de datos de columnas para 'True' rfe_support

col=X_train.columns[rfe.support_]
col
```

Usaremos statsmodel para construir nuestro modelo.

```python
# Creación de un nuevo conjunto de datos X después de aplicar rfe

X_train_rfe=X_train[col]
```

```python
# Creando un modelo lineal constante y en ejecución

X_train_const=sm.add_constant(X_train_rfe)

lm=sm.OLS(y_train,X_train_const).fit()

lm.summary()
```

## Creando VIF

```python
# Creando VIF

X_train_new=X_train_const.drop(['const'],axis=1)  

vif=pd.DataFrame()
X=X_train_new
vif['Predictors']=X.columns
vif['VIF']=[variance_inflation_factor(X.values,i) for i in range(X.shape[1])]
vif['VIF']=round(vif['VIF'],2)
vif=vif.sort_values(by='VIF',ascending=False)
vif
```
